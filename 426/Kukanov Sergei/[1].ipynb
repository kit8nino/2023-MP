```python
import datetime
from collections import deque

queue = deque()

#===============================Основной блок, ввод начальных данных================================
# Ввод данных с клавиатуры
fio = input("Введите ФИО: ")
birthdate = input("Введите дату рождения в формате ДД.ММ.ГГГГ: ")
# Спросить пользователя о способе ввода предметов
print("Выберите способ ввода предметов:")
print("1 - Ввести предметы вручную")
print("2 - Использовать готовый словарь и добавлять предметы (оценки выставляются заранее в коде!)")
choice = input()
#====================================================================================================

#================Код для выбора пресета школьных предметов или самостоятельного ввода================
if choice == "1":
    # Ввод предметов вручную
    subjects = {}
    while True:
        subject_name = input("Введите название предмета или 'стоп', чтобы закончить ввод: ")
        if subject_name == "стоп":
            break
        subject_score = input(f"Введите оценку за предмет '{subject_name}': ")
        subjects[subject_name] = int(subject_score)

elif choice == "2":
    # Готовый словарь предметов и возможность добавления новых предметов
    subjects = {"математика": 5, "русский язык": 4, "английский язык": 5, "физика": 4, "химия": 5,
                "биология": 5, "история": 4, "обществознание": 5, "литература": 4,
                "информатика": 5, "физическая культура": 5}
    print(f"Доступные предметы: {', '.join(subjects.keys())}")
    while True:
        add_subject = input("Хотите добавить новый предмет? (да/нет): ")
        if add_subject == "да":
            subject_name = input("Введите название нового предмета: ")
            subject_score = input(f"Введите оценку за предмет '{subject_name}': ")
            subjects[subject_name] = int(subject_score)
        elif add_subject == "нет":
            break

else:
    print("Неверный выбор. Попробуйте еще раз.")
    exit()
#====================================================================================================    

#==============Реализация FIFO очереди для предметов по вводимому индексу (+ названию)===============
while True:
    item = input("Введите предмет для FIFO очереди (или 'стоп' для завершения): ")
    if item == "стоп":
        break

    try:
        index = int(input("Введите индекс для вставки: "))
        queue.insert(index, item)
    except ValueError:
        queue.append(item)
#====================================================================================================          
        
#=====================Родственники, и функция, выцепляющая только имя из fio=========================
def extract_name(fio):
    # Разделить строку на отдельные слова
    words = fio.split()
    
    # Извлечь имя по индексу
    name = words[1]
    
    return name

name = extract_name(fio)

# Словарь с датами рождения родственников (со списком не удалось реализовать возможность ввода дат рождения)
relatives = {}

while True:
    relative_name = input("Введите имя родственника (или 'стоп' для выхода): ")
    if relative_name == "стоп":
        break
    
    relative_birthdate = input("Введите дату рождения родственника в формате ДД.ММ.ГГГГ: ")
    relatives[relative_name] = relative_birthdate
#==================================================================================================== 


pet_name = input("Введите имя своей пушистой кивы: ")

#====================================================================================================
#====================================Блок дополнительных заданий===================================== 
#==================================================================================================== 
# Разбиение даты рождения на составляющие
birthdate_parts = birthdate.split(".")
birth_day = int(birthdate_parts[0])
birth_month = int(birthdate_parts[1])
birth_year = int(birthdate_parts[2])
# 1) Средняя оценка в аттестате
average_score = sum(subjects.values()) / len(subjects)
print('=====================================================================================================================')
print(f"Средняя оценка в аттестате: {average_score}")

# 2) Уникальные имена родственников и их даты рождения, используя словарь, где ключ, имя родственника
unique_names = set(relatives.keys())  # создаем множество ключей словаря relatives
unique_names.add(name)  # добавляем имя самого человека в множество
print('=====================================================================================================================')
print(f"Уникальные имена родственников: {', '.join(unique_names)}")

# 3) Общая длина всех названий предметов
total_length = sum([len(subject) for subject in subjects.keys()])
print('=====================================================================================================================')
print(f"Общая длина всех названий предметов: {total_length}")

# 4) Уникальные буквы в названиях предметов
unique_letters = set()
for subject in subjects.keys():
    for letter in subject:
        unique_letters.add(letter)
print('=====================================================================================================================')
print(f"Уникальные буквы в названиях предметов: {', '.join(unique_letters)}")

# 5) Имя питомца в бинарном виде
binary_name = ''.join(format(ord(letter), '08b') for letter in pet_name)
print('=====================================================================================================================')
print(f"Имя пушистой кивы в бинарном виде: {binary_name}")

# 6) Отсортированный по алфавиту (в обратном порядке) список родственников
print('=====================================================================================================================')
print("Отсортированный список родственников в обратном алфавитном порядке:")
for name in sorted(relatives.keys(), reverse=True):
    print(name)

# 7) Количество дней от даты рождения до текущей даты
birth_date = datetime.date(birth_year, birth_month, birth_day)
today = datetime.date.today()
days_since_birth = (today - birth_date).days
print('=====================================================================================================================')
print(f"Количество дней от даты рождения до текущей даты: {days_since_birth}")

# 8) Вывод FIFO очереди
print('=====================================================================================================================')
print("FIFO Очередь:")
for item in queue:
    print(item)
if len(queue) == 0:
    print("FIFO Очередь пуста =)")

    
# 9) Ацтеки
print('=====================================================================================================================')
# Определение номера правителя
aztec = (birth_day + birth_month**2 + birth_year) % 21 + 1
# Список имен правителей
ruler_names = [
    "Acamapichtli", "Huitzilihuitl", "Chimalpopoca", "Itzcoatl", "Moctezuma I",
    "Axayacatl", "Tizoc", "Ahuitzotl", "Moctezuma II", "Cuitláhuac",
    "Cuauhtémoc", "Coanacoch", "Totoquihuatzin", "Quinatzin", "Ixtlilxochitl",
    "Nezahualcoyotl", "Maxtla", "Tlacaelel", "Tlacopan", "Tlatoani Diego"
]

# Замена имени в словаре
sorted_relatives = sorted(relatives.items())
if aztec <= len(sorted_relatives):
    relative_name_to_replace = sorted_relatives[aztec-1][0]
    ruler_name = ruler_names[aztec-1]
    del relatives[relative_name_to_replace]
    relatives[ruler_name] = sorted_relatives[aztec-1][1]
    print(f"Имя родственника '{relative_name_to_replace}' заменено на '{ruler_name}'")
else:
    print(f"Нет родственника с индексом {aztec} в списке, на этот раз без ацтеков")

print('----------------------------------------------------------------------------------------------------------------------')
print("Новый (или не тронутый) словарь с родственниками:")
print(relatives)

# 11) Функции-генераторы и тд тп
print('=====================================================================================================================')
# Определение значения number по формуле
number = (len(fio) * len(relatives) % 4)

# Функция-генератор аликвотной последовательности:
def aliquot_sequence(number):
    a = 0
    for i in range(10):
        yield a
        a = a + number
        if a == 0:
            break
            
# Функция-генератор последовательности Сильвестра:     
def sylvester_sequence():
    a = 2
    yield a
    for i in range(9):
        a = a * a + 1
        yield a
        
# Функция-генератор чисел трибоначчи:        
def tribonacci_sequence():
    a, b, c = 0, 0, 1
    yield a
    yield b
    yield c
    for i in range(7):
        d = a + b + c
        yield d
        a = b
        b = c
        c = d

# Функция-генератор чисел Леонардо:       
def leonardo_sequence():
    a, b = 1, 1
    yield a
    yield b
    for i in range(8):
        a, b = b, a + b + 1
        yield b
        
if number == 0:
    sequence = aliquot_sequence(len(fio) * len(relatives) % 4)
    sequence_name = "аликвотной последовательности"
elif number == 1:
    sequence = sylvester_sequence()
    sequence_name = "последовательности Сильвестра"
elif number == 2:
    sequence = tribonacci_sequence()
    sequence_name = "числа трибоначчи"
else:
    sequence = leonardo_sequence()
    sequence_name = "числа Леонардо"

print(f"Была выбрана функция-генератор {sequence_name}:")
for i, num in enumerate(sequence):
    print(num)
    if i >= 9:
        break
```
